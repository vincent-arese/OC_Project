```python
def StandardScaler_ML(df, features_list, LabelColString, standardization=True):
    global df_Source, features, data, data_scaled, LabelsArray, data_unscaled ,data_scaled_df 

    df_Source = df.copy()
    features = features_list
    data = df_Source[features]
    print(pcolors.OK +"Features Selection : "+pcolors.RESET)
    print(f"DataFrame source: {df_Source.shape}")
    print(f"Selected features: {len(features)}, {features}")
    
    LabelsArray = df_Source[LabelColString]
    print(f"LabelColString: {LabelColString}, not scaled, not fitted")
    print(f"LabelsArray: {LabelsArray}, not scaled, not fitted")
    
    diff = df_Source.columns.difference(features)
    print(f"Excluded features from analysis: {diff}")

    #### Preproscessing Standard Scalling 
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler(with_std=standardization)
    data_scaled = scaler.fit_transform(data)
    data_scaled_df = pd.DataFrame(data_scaled, columns=features)

    print(f"\n{pcolors.OK}Preprocessing : Scaling :{pcolors.RESET} {scaler}")
    print(f"Data scaled from {df_Source.shape} to {data_scaled.shape}")
    print(f"\n{pcolors.WARNING}Globals Variables :{pcolors.RESET}")
    print(f"df_Source: source DataFrame (before feature selection)")
    print(f"LabelColString ({LabelColString} ): label column, not fitted")
    print(f"features = features_list : selected features")
    print(f"data: DataFrame with selected features")
    print(f"data_scaled: data_scaled Array with StandardScaler")
    print(f"{pcolors.BLUE}data_scaled_df :{pcolors.RESET} data_scaled DataFrame with StandardScaler\n")
    
    #INFOS(data_scaled_df)
    print(pcolors.WARNING+"describes data_scaled_df:"+pcolors.RESET,"\n")
    print(data_scaled_df.describe().T.to_markdown()) 
    print()
    return  data_scaled_df
```

###<font color="#00d16c"> Standard Scaler recherche Outliers DBSCAN</font> </a>

```python
def DBSCANoutliers(df_source, cluster_min_samples):
    from sklearn.preprocessing import StandardScaler
    from sklearn.cluster import DBSCAN

    # Charger les données
    DBSCANdata = df_source.copy() 
    
    # Standardiser les données
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(DBSCANdata)

    # Clustering DBSCAN 
    min_distance_eps = 0.3 # Avec une distance inférieure à 0.3
    dbscan = DBSCAN(eps=min_distance_eps, min_samples=cluster_min_samples) 
    dbscan.fit(data_scaled)
    DBSCANdata["DBSCAN_Cluster_Label"] = dbscan.labels_
 

    # Trouver les outliers
    outliers = DBSCANdata[DBSCANdata["DBSCAN_Cluster_Label"] == -1].index
    DBSCANdata.loc[outliers, "DBSCAN_Cluster_Label"] = "Outliers"

    # Afficher les labels de clustering
    print(f"Les labels de clustering : {DBSCANdata.DBSCAN_Cluster_Label.unique()}")

    outliers_query = DBSCANdata[DBSCANdata.DBSCAN_Cluster_Label == "Outliers"]
    outliers_percent = (len(outliers_query) / DBSCANdata.shape[0]) * 100
    print(f"Il y a {len(outliers_query)} soit {outliers_percent :.2f}% d'outliers dans le dataframe {DBSCANdata.shape}")

    return DBSCANdata
```

```python
def RobustScaler_ML(df, features_list, LabelColString, robust=True):
    global df_Source, features, data, data_scaled, LabelsArray, data_unscaled ,data_scaled_df 

    df_Source = df.copy()
    features = features_list
    data = df_Source[features]
    print(pcolors.OK +"Features Selection : "+pcolors.RESET)
    print(f"DataFrame source: {df_Source.shape}")
    print(f"Selected features: {len(features)}, {features}")
    
    LabelsArray = df_Source[LabelColString]
    print(f"LabelColString: {LabelColString}, not scaled, not fitted")
    print(f"LabelsArray: {LabelsArray}, not scaled, not fitted")
    
    diff = df_Source.columns.difference(features)
    print(f"Excluded features from analysis: {diff}")

    #### Preproscessing Robust Scaling 
    #https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html
    from sklearn.preprocessing import RobustScaler
    scaler = RobustScaler(with_centering=robust)
    data_scaled = scaler.fit_transform(data)
    data_scaled_df = pd.DataFrame(data_scaled, columns=features)

    print(f"\n{pcolors.OK}Preprocessing :{pcolors.RESET} Scaling  {scaler}")
    print(f"Data scaled from {df_Source.shape} to {data_scaled.shape}")
    print(f"\n{pcolors.WARNING}Globals Variables :{pcolors.RESET}")
    print(f"df_Source: source DataFrame (before feature selection)")
    print(f"LabelColString ({LabelColString} ): label column, not fitted")
    print(f"features = features_list : selected features")
    print(f"data: DataFrame with selected features")
    print(f"data_scaled: data_scaled Array with RobustScaler")
    print(f"{pcolors.BLUE}data_scaled_df :{pcolors.RESET} data_scaled DataFrame with RobustScaler\n")
    
    #INFOS(data_scaled_df)
    print(pcolors.WARNING+"describes data_scaled_df:"+pcolors.RESET,"\n")
    print(data_scaled_df.describe().T.to_markdown()) 
    print()

    return  data_scaled_df
```

```python
def ACP_ML(data_scaled):
    global PCA_component_df , pca_fited, PCA_component_array , eigenvalues, listePCName,PCA_n_components, X_projected, PCA_exp_var 
    # ACP: Analyse en Composantes Principales
    print(pcolors.OK + "\nACP : Analyse en Composantes Principales" + pcolors.RESET)

    print(f"data_scaled shape : {data_scaled.shape}")

    from sklearn.decomposition import PCA

    X_scaled = data_scaled    # Dataframe Scaled si nécessaire
    n_components = X_scaled.shape[1]   # nombre de compantes principales
    pca = PCA(n_components=n_components)
    pca_fited = pca.fit(X_scaled) #Modele entrainée

    print(f"{pcolors.WARNING} pca.explained_variance_ratio:{pcolors.RESET}\n {pca.explained_variance_ratio_}")
    PCA_component = pca.components_
    PCA_component_array = PCA_component
    print(f"PCA_component_array.shape : {PCA_component_array.shape} ")

    listePCName = [f'PC{i}' for i in range(1, PCA_component.shape[0] + 1)]
    PCA_n_components = n_components
    print(f"PCA_ n_components : {PCA_n_components} ")
    print(f"listePCName  : {listePCName} \n")


    X_projected = pca.transform(X_scaled)
    print(f"X_projected.shape : {X_projected.shape}\n")
    
    pcs = pd.DataFrame(PCA_component)
    pcs.columns = listePCName
    pcs["features"] = features
    pcs = ColMove(pcs, "features", 0)

    PCA_exp_var = pca.explained_variance_ratio_ #Explained Variance des Composantes Principales
    print(f"pca.explained_variance_ratio_ : {pca.explained_variance_ratio_}")

    eigenvalues = (n_components-1)/n_components*pca.explained_variance_  # eigenvalues : Les valeurs propres en fonction des facteurs
    print(f"eigenvalues : {eigenvalues} (baton brisés ) \n")

    print(f"{pcolors.BLUE}PCA_component_df{pcolors.RESET}")

    PCA_component_df = pcs

    return PCA_component_df

# ACP V4  MAJ 29/01/23
# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
# https://sites.google.com/view/aide-python/statistiques/machine-learning-en-python/analyses-en-composantes-principale
# https://blog.statoscop.fr/acp-python.html 
```

```python
def BatonsBrises():
 print( f"\n{pcolors.OK}Méthode des bâtons brisés {pcolors.RESET} : {pca_fited}")
 # Methode des batons brisés : Calcul du seuil theorique de l'axe 
 # Ne conserver que les axes sont l'inertie  est >= au seuil théorique de l'axe 

 n_components = len(features)

 b_i = 1/np.arange(n_components,0,-1)
 cumul_b_i = np.cumsum(b_i)
 cumul_b_i = cumul_b_i[::-1]
 BatonsBrise = pd.DataFrame({'ComposantePrincipale':listePCName ,'Val_Propre_eigen':eigenvalues,'Seuil':cumul_b_i})
 BatonsBrise["BatonBrise"] = np.where(BatonsBrise['Val_Propre_eigen']>=BatonsBrise['Seuil'], 'Axe_interessant ', 'Axe_Secondaire')
 print("Axe intéressant  : Valeur Propre Axe eigen>=Seuil  - Axe secondaire  : eigen<Seuil théorique")
 print("")
 print(BatonsBrise.to_markdown())
 #return BatonsBrise
```

```python
# Explained Variance & Scree Plot => Plotly Version
def ACP_EboulisPlot():

   # pca_fited, PCA_n_components
    print( f"\n{pcolors.OK}Eboulis des valeurs Propres {pcolors.RESET} : {pca_fited}") 

    # Gather the explained variance by each Principal Component and accumulate them
    PCA_exp_var_percent = (PCA_exp_var *100).round(2)
    exp_var_cumul = (np.cumsum(PCA_exp_var_percent)).round(2)

    # Number of Principal Components analyzed in ACP
    n_components = PCA_n_components
    x_list = list(range(1, n_components+1))

    # Plot the explained variance
    titleString = f"{df_Source_Name_String} : Explained Variance of Principal Components"
    fig = px.bar(x=x_list, y=PCA_exp_var_percent,
                 barmode='group',
                 orientation="v",
                 opacity=1,
                 labels=dict(x="Component", y="Variance % Inertia"),
                 text_auto=True,
                 title=titleString ,
                 width=1000, height=500)

    # Plot the cumulative explained variance
    fig0 = px.line(x=x_list, y=exp_var_cumul,
                   labels=dict(x="Component", y="Cumulative Variance % Inertia"),
                   title='Cumulative Variance',
                   markers=True)
    fig0.update_traces(line_color='red',
                       fillcolor='rgba(255, 0, 0, 0.1)',
                       line_width=3,
                       mode='lines+markers',
                       marker_size=15)

    # Add the cumulative variance plot to the explained variance plot
    fig.add_trace(fig0.data[0])

    # Add markers
    fig.add_hline(y=100, line_width=1, line_dash="dash", line_color="grey")
    fig.add_hline(y=80, line_width=1, line_dash="dash", line_color="grey")

    # Remove the legend
    fig.update_traces(showlegend=False)

    # sauvergardr le graphique en Html 
    PATH = f"/content/Exports/{df_Source_Name_String}/"
    if not os.path.exists(PATH):
     os.makedirs(PATH)
    file_name = f"{PATH}{df_Source_Name_String}.html"
    ExportPlotly(fig,file_name)
    print(f"Link to the file: {file_name}")   

    return fig
```
